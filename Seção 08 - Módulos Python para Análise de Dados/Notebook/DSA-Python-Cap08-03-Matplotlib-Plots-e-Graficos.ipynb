# Construindo Plots

```python
# O matplotlib.pyplot é uma coleção de funções e estilos que fazem com que o Matplolib funcione como o Matlab.
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mpl.__version__
```

```python
# O método plot() define os eixos do gráfico
plt.plot([1, 2, 3], [2, 5, 7])
plt.show()
```

```python
x = [1, 4, 5]
```

```python
y = [3, 7, 2]
```

```python
plt.plot(x, y)
plt.xlabel('Variável 1')
plt.ylabel('Variável 2')
plt.title('Teste Plot')
plt.show()
```

```python
x2 = [1, 2, 3]
y2 = [11, 12, 15]
```

```python
plt.plot(x2, y2, label = 'Primeira Linha')
plt.legend()
plt.show()
```
